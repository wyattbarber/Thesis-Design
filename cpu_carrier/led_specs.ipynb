```python
COLORS = {
    "PWR": "RED",
    "ACT": "GREEN",
    "APP0": "YELLOW",
    "APP1": "RED"
}
V_FORWARD = {
    "RED": 1.9,
    "GREEN": 2.8,
    "YELLOW": 1.8
}
V_SUPPLY = 3.3
I_MAX = 0.01
```

```python
for signal, color in COLORS.items():
    Vr = V_SUPPLY - V_FORWARD[color]
    R = Vr / I_MAX
    print(f"{signal}: {R} Ohms, {Vr} Volts")
```
